```python
# Renombramos la columna "id" en movies a "movie_id" para que coincidan
movies.rename(columns={"id": "movie_id"}, inplace=True)

# Realizamos el merge usando la columna "movie_id"
merged_df = pd.merge(movies, credits, on="movie_id", how="inner")

# Mostramos los primeros registros del DataFrame unificado
merged_df.columns
merged_df.head()
```

```python
# Seleccionar las columnas especÃ­ficas con .loc
df = merged_df.loc[:, ['movie_id', 'title_x',"title_y", 'overview', 'genres', 'keywords', 'cast', 'crew']]


# Verificar si las columnas "title_x" y "title_y" tienen los mismos datos
print(df['title_x'].equals(df['title_y']))
#si sin iguales devuelve true
```

```python
# Eliminamos una de las columnas y renombramos a "title" la otra
df=df.drop("title_y",axis=1)
df.rename(columns={"title_x": "title"}, inplace=True)
df.head()
```

```python
# FunciÃ³n para extraer la lista de 'name' de un JSON y reemplazar la columna
def extract_names(json_str):
    try:
        # Convierte la cadena en un objeto de Python
        json_obj = ast.literal_eval(json_str)
        # Extrae el atributo 'name' de cada diccionario
        return ', '.join([item['name'] for item in json_obj])  # Combina los gÃ©neros como una cadena
    except (ValueError, SyntaxError):
        return ''

# FunciÃ³n para obtener los tres primeros nombres del reparto
def extract_top_cast(json_str):
    try:
        json_obj = ast.literal_eval(json_str)
        return ', '.join([item['name'] for item in json_obj[:3]])  # Combina los tres nombres como una cadena
    except (ValueError, SyntaxError):
        return ''
```

```python
# Transformar y reemplazar las columnas
df['genres'] = df['genres'].apply(extract_names)  # Reemplaza 'genres' por los nombres de los gÃ©neros
df['keywords'] = df['keywords'].apply(extract_names)  # Reemplaza 'keywords' por los nombres de las palabras clave
df['cast'] = df['cast'].apply(extract_top_cast)  # Reemplaza 'cast' por los tres primeros nombres del reparto
df.head()
```

```python
# FunciÃ³n para obtener el nombre del director de la columna 'crew'
def get_director(json_str):
    try:
        json_obj = ast.literal_eval(json_str)  # Convierte el string JSON en un objeto Python
        for item in json_obj:
            if item['job'] == 'Director':  # Busca el director en el equipo
                return item['name']  # Devuelve el nombre del director
        return None  # Si no se encuentra un director
    except (ValueError, SyntaxError):
        return None

# FunciÃ³n para convertir 'overview' (resumen) en una lista de palabras
def convert_overview_to_list(text):
    try:
        return text.split()  # Convierte el texto en una lista de palabras
    except AttributeError:  # Maneja valores nulos o no vÃ¡lidos
        return []

# Procesar la columna 'crew' para que solo contenga el nombre del director
df['crew'] = df['crew'].apply(get_director)

# Procesar la columna 'overview' para convertirla en una lista de palabras
df['overview'] = df['overview'].apply(convert_overview_to_list)

df.head()
```

```python
# FunciÃ³n para quitar los espacios en los elementos de una lista o cadenas de texto
def remove_spaces(value):
    if isinstance(value, list):  # Si el valor es una lista
        return [item.replace(' ', '') for item in value]  # Elimina los espacios de cada elemento
    elif isinstance(value, str):  # Si el valor es una cadena de texto
        return value.replace(' ', '')  # Elimina los espacios
    else:
        return value  # Devuelve el valor sin cambios si es otro tipo

# Aplicar la funciÃ³n a las columnas correspondientes
df['genres'] = df['genres'].apply(remove_spaces)
df['cast'] = df['cast'].apply(remove_spaces)
df['crew'] = df['crew'].apply(remove_spaces)
df['keywords'] = df['keywords'].apply(remove_spaces)
```

```python
# FunciÃ³n para limpiar espacios y comillas extras
def clean_string(value):
    # Eliminar espacios extras y comillas no deseadas
    return ' '.join(str(value).replace("'", "").split())

# FunciÃ³n para combinar todas las columnas en una sola cadena
def combine_columns(row):
    # Limpiar y combinar los valores de todas las columnas
    return ' '.join([
        clean_string(row['overview']),
        clean_string(row['genres']),
        clean_string(row['keywords']),
        clean_string(row['cast']),
        clean_string(row['crew'])
    ])

# Crear la nueva columna 'tags' combinando las columnas procesadas
df['tags'] = df.apply(combine_columns, axis=1)

# Reemplazar comas por espacios en blanco en la columna 'tags'
df['tags'] = df['tags'].str.replace(',', '')

# Mostrar el primer valor de la columna 'tags' para verificar
df["tags"][0]
```

```python
df["tags"] = df["tags"].str.strip().str.lower()
df['tags'] = df['tags'].str.replace(" ", "")

vectorizer = TfidfVectorizer()
df_vectorized = vectorizer.fit_transform(df["tags"])
print(df_vectorized)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(df_vectorized)
```

```python
model = NearestNeighbors(n_neighbors = 6, metric = "cosine")
model.fit(df_vectorized)
```

```python
def recommend(movie):
    movie_index = df[df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
def recommend(movie):
    # Convertir los tÃ­tulos a minÃºsculas para comparaciÃ³n insensible a mayÃºsculas
    movie = movie.lower()
    df['title_lower'] = df['title'].str.lower()  # Crear una columna auxiliar en minÃºsculas
    
    # Verificar si la pelÃ­cula existe en el DataFrame
    if movie not in df['title_lower'].values:
        print("PelÃ­cula no encontrada. Por favor, verifica el tÃ­tulo.")
        return

    # Obtener el Ã­ndice de la pelÃ­cula consultada
    movie_index = df[df["title_lower"] == movie].index[0]

    # Calcular distancias de similitud
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # Mostrar recomendaciones
    print(f"PelÃ­culas similares a '{df.iloc[movie_index].title}':")
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend("avatar")
```
